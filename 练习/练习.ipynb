猜数游戏。在程序中预设一个0~100之间的整数，让用户通过键盘输入所猜的数，如果大于预设的数，显示“遗憾，太大了”；小于预设的数，显示“遗憾，太小了”，如此循环，直至猜中该数，显示“预测N次，你猜中了！”，其中N是用户输入数字的次数。‪‬‪‬‪‬‪‬‪‬‮‬‪‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‪‬‭‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‭‬‫‬

提示： 生成随机数可以使用以下方法：‪‬‪‬‪‬‪‬‪‬‮‬‪‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‪‬‭‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‭‬‫‬

import random‪‬‪‬‪‬‪‬‪‬‮‬‪‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‪‬‭‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‭‬‫‬

Set_number = random.randint(0,100)

````python
import random
Set_number = random.randint(0,100)
N = 0
while True:
    N += 1
    n = int(input())
    if (n>Set_number):
        print('遗憾，太大了')
    elif (n<Set_number):
        print('遗憾，太小了')
    else:
        print('预测%d次，你猜中了！'%N)
        break
````

(1) 输入输出的摄氏度采用大写字母C开头，温度可以是整数或小数，如：C12.34指摄氏度12.34度；‪‬‪‬‪‬‪‬‪‬‮‬‪‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‪‬‭‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‭‬‫‬

(2) 输入输出的华氏度采用大写字母F开头，温度可以是整数或小数，如：F87.65指摄氏度87.65度；‪‬‪‬‪‬‪‬‪‬‮‬‪‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‪‬‭‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‭‬‫‬

(3) 不考虑异常输入的问题，输出保留小数点后两位；‪‬‪‬‪‬‪‬‪‬‮‬‪‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‫‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‪‬‭‬‪‬‪‬‪‬‪‬‪‬‮‬‫‬‮‬‪‬‪‬‪‬‪‬‪‬‮‬‭‬‫‬

(4) 使用input()获得测试用例输入时，不要增加提示字符串。

````python
a = input()
if (a[0]=='F'):
    c = (float(a[1:])-32)/1.8
    print('C%.2f'%c)
elif (a[0]=='C'):
    f = float(a[1:])*1.8+32
    print('F%.2f'%f)
````

4.3
蒙特卡罗方法近似计算圆周率值


````python
from random import random
DARTS =1000
hits = 0
for i in range(DARTS):
    x,y = random(),random()
    dist = pow(x**2+y**2,0.5)
    if dist <= 1.0:
        hits +=1
pi = 4*(hits/DARTS)
print("{}".format(pi))
````

````python
import csv
file = open('iris.csv', 'r')
reader = csv.DictReader(file)
v = []
s = []
for i in reader:
    if i['species'] == 'versicolor':
        v.append(float(i['petal_width']))
    elif i['species'] == 'setosa':
        s.append(float(i['petal_width']))
file.close()
if len(v) > 0 and len(s) > 0:
    avg_v = sum(v) / len(v)
    avg_s = sum(s) / len(s)
    print(f'{avg_v - avg_s:.1f}')  
    print(round(avg_v - avg_s, 1))  
````

````python
import csv

# 使用 open 打开文件
csvfile = open('iris.csv', newline='')  # 打开文件
reader = csv.DictReader(csvfile)

# 用于存储 versicolor 和 setosa 的 petal_width
v_widths = []
s_widths = []

# 遍历每一行数据
for row in reader:
    if row['species'] == 'versicolor':
        v_widths.append(float(row['petal_width']))
    elif row['species'] == 'setosa':
        s_widths.append(float(row['petal_width']))

# 计算平均值并返回差值
if len(v_widths) > 0 and len(s_widths) > 0:
    avg_v = sum(v_widths) / len(v_widths)
    avg_s = sum(s_widths) / len(s_widths)
    csvfile.close()  # 记得关闭文件
    print(round(avg_v - avg_s, 1))
else:
    csvfile.close()  # 关闭文件
    print("NOT FOUND")
````
